```python
# Creating columns for precipitation categories (precipitable_water_entire_atmosphere) 
# light < 2.5, moderate 2.5 - 10.0, heavy 10.0 - 50.0, violent > 50.0
train['prep_cat'] = ['light' if 0<precipitable_water_entire_atmosphere<2.5 else 'moderate' \
                      if 2.4<precipitable_water_entire_atmosphere<10.0 else 'heavy' if 9.9<precipitable_water_entire_atmosphere<50.0 \
                      else 'violent' if 49.9<precipitable_water_entire_atmosphere<74.0 else None \
                      for precipitable_water_entire_atmosphere in list(train['precipitable_water_entire_atmosphere'].values)]

test['prep_cat'] = ['light' if 0<precipitable_water_entire_atmosphere<2.5 else 'moderate' \
                      if 2.4<precipitable_water_entire_atmosphere<10.0 else 'heavy' if 9.9<precipitable_water_entire_atmosphere<50.0 \
                      else 'violent' if 49.9<precipitable_water_entire_atmosphere<74.0 else None \
                      for precipitable_water_entire_atmosphere in list(test['precipitable_water_entire_atmosphere'].values)]
```

```python
# creating a labelEncoder object
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder() #Bound labelEncoder to le to reduceName
#transforming prep_cat which return light,moderate,heavy and violent into numerical values
train['prep_cat'] = le.fit_transform(train['prep_cat'])
test['prep_cat'] = le.fit_transform(test['prep_cat'])
```

```python
# Dropping columns with lots of missing data and those those not present in test data 
train.drop(['Place_ID X Date','Date','Place_ID','target_min','target_max','target_variance','target_count','L3_CH4_CH4_column_volume_mixing_ratio_dry_air',
       'L3_CH4_aerosol_height', 'L3_CH4_aerosol_optical_depth',
       'L3_CH4_sensor_azimuth_angle', 'L3_CH4_sensor_zenith_angle',
       'L3_CH4_solar_azimuth_angle', 'L3_CH4_solar_zenith_angle'], axis=1, inplace=True)
test.drop(['Place_ID X Date','Date','Place_ID', 'L3_CH4_CH4_column_volume_mixing_ratio_dry_air',
       'L3_CH4_aerosol_height', 'L3_CH4_aerosol_optical_depth',
       'L3_CH4_sensor_azimuth_angle', 'L3_CH4_sensor_zenith_angle',
       'L3_CH4_solar_azimuth_angle', 'L3_CH4_solar_zenith_angle'], axis=1, inplace=True)
```

If you don't know about creating lag features, check these articles :
https://www.mikulskibartosz.name/forecasting-time-series-using-lag-features/
https://www.analyticsvidhya.com/blog/2019/12/6-powerful-feature-engineering-techniques-time-series/

```python
lag = ['relative_humidity_2m_above_ground', 'v_component_of_wind_10m_above_ground',
        'L3_NO2_NO2_column_number_density', 'L3_HCHO_tropospheric_HCHO_column_number_density',
        'L3_CLOUD_surface_albedo', 'L3_CO_CO_column_number_density',
        'temperature_2m_above_ground', 'L3_CLOUD_cloud_optical_depth', 'L3_O3_O3_column_number_density'
      ]

for features in lag:
    for i in range(3):
        train[features+'_lag'+str(i+1)] = train[features].shift(i+1)
        test[features+'_lag'+str(i+1)] = test[features].shift(i+1)
```

```python
import datasist as ds # Using datasist library to fill missing numerical data in the data
# The fill_missing_num uses the mean to fill data
train = ds.feature_engineering.fill_missing_num(train)
test = ds.feature_engineering.fill_missing_num(test)
```

## Using different models to find the optimal model
Just trying out different things here

Documentation that can help here:
https://catboost.ai/docs/concepts/python-usages-examples.html#cross-validation
https://lightgbm.readthedocs.io/en/latest/Parameters.html
https://neptune.ai/blog/lightgbm-parameters-guide

```python
# Note: Running this code will take some time to complete
# Using different models to find the optimal model
#
from sklearn.model_selection import KFold, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error 
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from catboost import CatBoostRegressor


# Creating a list of regressor algorithms to compare with
#
models = [RandomForestRegressor(), 
          LGBMRegressor(objective= 'regression', boosting= 'gbdt', metric= 'rmse',
                        learning_rate= 0.05, num_iterations= 7500, max_depth= -1,
                        min_data_in_leaf= 20, bagging_fraction= 0.8, bagging_freq= 1,
                        feature_fraction= 0.8, verbose_eval=500), 
         XGBRegressor(objective ='reg:squarederror'), 
         LinearRegression(), 
         CatBoostRegressor(min_data_in_leaf= 20, logging_level='Silent')]


# Creating lists of the algorithms, to store the accuracy scores of each fold
#
RandomForest, LGB, XGB, Linear, Cat = ([] for x in range(5))


# Creating a list containig the list of each algorithm. Created for easy iteration
#
model_list = [RandomForest, LGB, XGB, Linear, Cat]


# Spliting the data into features and the target variable
#
X = train.drop(['target'], axis = 1)
y = train.target


# Creating a cross validation of 10 folds
#
kfold  = KFold(n_splits=10, shuffle=True, random_state=42)


# Iterating through each model and appending the scores of each fold to the appriopriate list
#
for i, j in zip(models, model_list):
  j.extend(list(cross_val_score(i, X, y, scoring = 'neg_mean_squared_error', cv = kfold)))

  
# Creating a function to convert neg_mean_squared_error to a square root
#
def sq(lis):
  new_lis = []
  lis = np.array(lis)
  for i in lis:
    i = np.sqrt(i*-1)
    new_lis.append(i)
  return new_lis


# Creating a dataframe of all the rmses from the iterations for each model
#
rmses = pd.DataFrame({'Fold': np.arange(1, 11), 'RandomForest': sq(RandomForest), 'LGB': sq(LGB),
                      'XGB': sq(XGB), 'Linear': sq(Linear), 'Cat': sq(Cat)})

# Setting the index
#
rmses.set_index('Fold', inplace = True)


# Calculating the mean and standard deviation rmse of each algorithm
#
rmses.loc['mean'] = rmses.mean()
rmses.loc['std'] = rmses.std()


# Previewing the rmses dataframe
#
rmses
```

```python
# Checking for the regressor with minimum root mean squared error
#
rmses.loc['mean'].idxmin(), rmses.loc['mean'].min()
```

```python
# Arranging the models in ascending order
#
rmses.loc['mean'].sort_values()
```

```python
# Using the top three models; LGB, Catboost and XGBoost to train and make predictions
# Creating a list of models to use
models = [LGBMRegressor(objective= 'regression',
         boosting= 'gbdt',  
         metric= 'rmse',
         learning_rate= 0.05, 
         num_iterations= 2200,
         max_depth= -1,
         min_data_in_leaf= 15,
         bagging_fraction= 0.5,
         bagging_freq= 1,
         feature_fraction= 0.8, verbose_eval=500),
          CatBoostRegressor(logging_level='Silent'),
          XGBRegressor(objective ='reg:squarederror')]
model_names = ['LGB', 'catboost', 'xgboost']


# Selecting the training features and the target feature
#
X = train.drop(['target'], axis = 1)
y = train['target']


# Submission dataset
#
sub = test


# Using a for loop to create a submission file for each model
#
for model, model_name in zip(models, model_names):
  regressor = model                      # instantiating the model
  regressor.fit(X, y)                    # Training the model
  predictions  = regressor.predict(sub)  # Making predictions
  submission_df = pd.DataFrame({'Place_ID X Date': sample['Place_ID X Date'], 'target': predictions}) # Creating a submission file
  submission_df.to_csv(model_name + '_prediction.csv', index = False)
```
